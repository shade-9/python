2. a)Write a Python programming to display a bar chart of the popularity of programming
Languages. Also draw Pie chart for popularity Data values.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
x = np.array(["Java", "Python", "PHP", "JavaScript", "C#", "C++"])
y = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

plt.title("Popularity of programming languages")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")

plt.bar(x, y, color = "#8DEEEE")
plt.show()

y = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
mylabels = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]

plt.title("Popularity of programming languages")

plt.pie(y, labels = mylabels)
plt.show()
```

b) Write a Python program to display a horizontal bar chart of the popularity of programming
Languages. Hint: use the barh() function

```python
x = np.array(["Java", "Python", "PHP", "JavaScript", "C#", "C++"])
y = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

plt.title("Popularity of programming languages")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")

plt.barh(x, y, color = "#4F930E")
plt.show()
```
